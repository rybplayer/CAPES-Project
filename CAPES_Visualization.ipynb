## Part 2: Whole Dataset Trends

Here, we will try to visualize trends across the entire dataset.

```python
Gpa_Exp = Capes.get('Avg GPA Expected')
Gpa_Rec = Capes.get('Avg GPA Received')

plt.hist(Gpa_Exp, alpha = 0.5, density=True, label='GPA Expected', bins=np.arange(2,4,0.1));
plt.hist(Gpa_Rec, alpha = 0.5, density=True, label='GPA Received', bins=np.arange(2,4,0.1));

plt.rcParams['figure.dpi'] = 300;
plt.title('GPA Expected vs. GPA Received')
plt.xlabel('GPA (0 - 4.0)')
plt.ylabel('Proportion')
plt.legend(loc='upper left')
```

```python
Capes_filter = Capes[-Capes.get('Course Type').str.contains('Seminar')]
Capes_filter.groupby('Study Hrs/wk').mean(numeric_only=True).get(['Avg GPA Expected','Avg GPA Received'])\
    .reset_index().plot(kind='scatter',x='Study Hrs/wk',y='Avg GPA Expected', figsize=[5,5])
```

```python
def GpaLinearRegression(name):
    Capes_filter = Capes[Capes.get('Course Name') == name]
    if Capes_filter.shape[0] != 0:
        Capes_groupby = Capes_filter.groupby('Term Year').mean(numeric_only=True)\
            .get(['Avg GPA Expected','Avg GPA Received']).reset_index()
        
        sns.regplot(x='Term Year',y='Avg GPA Expected',data=Capes_groupby,label='Avg GPA Expected')
        sns.regplot(x='Term Year',y='Avg GPA Received',data=Capes_groupby,label='Avg GPA Received')
        
        plt.legend()
```

```python
GpaLinearRegression('MATH 20D')
```
