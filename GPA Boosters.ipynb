```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import dataframe_image as dfi
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
Capes = pd.read_csv('CAPES_tidy.csv')
Capes = Capes.drop(columns=Capes.columns[0])
Capes 
```

```python
Prereq = pd.read_csv('prereq.csv')
Prereq = Prereq.drop(columns=Prereq.columns[0])
Prereq
```

```python
Merged = Prereq.merge(Capes, left_on='course', right_on='Course Name')
Merged
```

```python
numbers = re.compile(r'\d+(?:\.\d+)?')
```

```python
Boosters = Merged[(Merged.get('prereqs') == "[]") & 
                  (Merged.get('Avg GPA Received') >= 3.7) & 
                  (Merged.get('Term Year') > 21) & 
                  (Merged.get('Enroll') >= 50) &
                  (Merged.get('course').apply(numbers.findall).apply(lambda x: x[0]).apply(int) < 100) &
                  (Merged.get('Evals Made')*2 >= Merged.get('Enroll'))]
Boosters
```

```python
Unicorns = Boosters.groupby('course').mean(numeric_only=True).sort_values(by='Avg GPA Received',ascending=False)
Unicorns
```

```python
# dfi.export(Unicorns, 'Unicorns.png')
```

```python
PureUnicorns = Unicorns.drop(['DOC 1','WCWP 10A','DOC 2']).drop(columns=['Term Year'])
PureUnicorns
```
